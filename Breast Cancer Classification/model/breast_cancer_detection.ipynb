## Model

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Flatten,BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.model_selection import train_test_split
```

```python
!pip install keras-tuner
```

```python
import keras_tuner as kt
```

```python
def build_model(hp):
  no_of_layers = hp.Int('num_layers',min_value=2,max_value=5)
  model = Sequential()
  model.add(Flatten(input_shape=(30,)))
  for i in range(no_of_layers):
    no_of_neurons = hp.Int(f'num_nodes_layer{i}',min_value=32,max_value=64,step=8)
    model.add(Dense(no_of_neurons,activation='relu'))
  model.add(Dense(2,activation='sigmoid'))
  model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
  return model
```

```python
callback = EarlyStopping(
    monitor='val_loss',
    min_delta=0.01,
    verbose=1,
    patience=20,
    mode='auto'
)
```

```python
x = df.drop(['target'],axis=1)
y = df[['target']]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
tuner = kt.RandomSearch(build_model,objective='val_accuracy',max_trials=5,directory='keras-tuner',project_name='breast_cancer_prediction')
```

```python
tuner.search(x_train,y_train,epochs=20,validation_data=(x_test,y_test))
```

```python
tuner.get_best_hyperparameters()[0].values
```

```python
best_model = tuner.get_best_models()[0]
```

```python
best_model.fit(x_train,y_train,epochs=1000,callbacks=callback,validation_data=(x_test,y_test))
```

```python
test_1 = x_test.values[0].reshape(1,-1)
```
